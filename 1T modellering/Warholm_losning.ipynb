## Hva hadde verdensrekorden blitt med konstant fart fra 2. hekk?

Videre skal vi finne et bedre estimat for samme spørsmål ved hjelp av regning.

<div class="alert alert-block alert-info">
<b>Oppgave: </b>Bruk x og y verdiene som tilsvarer 1. og 2. hekk, finn stigningstall <b>a</b> og konstantledd <b>b</b> til linjen som går gjennom disse punktene. Løs likningen $a\cdot x+b=400$ for å svare på spørsmålet.<br><br>
<b>Utfordring:</b>Kan du løse samme oppgave  med koding? Skriv en kodesnutt som tar inn to xverdier og to tilhørende yverdier, bruker disse til å regne ut a, b og løsning på likingen. Programmet bør skrive ut svarene med passelig antall siffer og en tilhørende tekst. </div>

```python
y1=45
y2=80
t1=5.35
t2=8.9

a=(y2-y1)/(t2-t1) # stigningstall = delta_y/delta_t
b=y1-a*t1 #løser ett punktsformelen og ser hva som utgjør konstantleddet
sluttid=(400-b)/a # løser ax+b=400 for x

print(f'Stigningstall a={a:.2f} og konstantleddet b={b:.2f}, det gir en slutttid på t={sluttid:.2f}s')
```

## Farten gjennom løpet
Vi ønsker å lage plot som viser hvordan gjennomsnittsfaren varier gjennom løpet.
<div class="alert alert-block alert-info">
<b>Oppgave: </b>Skriv en forklaring til hva kodesnutten under gjør og legg til kommentarer etter hver av linjene</div>

```python
fart=[] #lager en tom liste til å samle opp fartsberegningene 

for i in range(1,len(tid)): # starter 1. ikke 0. element siden vi bruker i-1 i utregning og løper gjennom tidslisten
    v=(strekning[i]-strekning[i-1])/(tid[i]-tid[i-1]) # regener ut delta_s/delta_t for en og en måling
    fart.append(v) # legger verdien til slutt i listen: fart.
    
```

**svar:**
Denne kodesnutten løper gjennom listen for tid og regner ut gjennomsnitts farten mellom to og to målinger av posisjon.Hver ny utregning blir lagt til på slutten av listen *fart*. Siden vi trenger to målinger av posisjon for å regne ut en verdi for fart blir fartslisten ett element kortere enn tid og strekning - listene.

<div class="alert alert-block alert-info">

```python
tid_midt=[]
for i in range(1,len(tid)):
    tid_midt.append((tid[i-1]+tid[i])/2)
    
plt.plot(tid_midt,fart)
plt.xlabel('tid/$s$')
plt.ylabel('fart/$m/s$')
plt.title('Fart - 400m hekk') 
plt.show()
```

## Ved hvilket tidspunkt skal man plotte en gjennomsnittsfart?

For å plotte farten over valgte du nok en av fire løsninger: legge til farten = 0 i starten av fartslisten, ta bort første tidsmåling, ta bort siste tidsmåling fra listen eller regne ut gjennomsnittet av to og to tidspunkt og lage en ny tidsliste. 

Det fins en femte løsning: tegne et stegplott for å vise at verdiene gjelder for ett og ett intervall.

## Kurvetilpassing
Nå har du laget modeller av ol-løpet som punkter i en graf og trukket rett linjer mellom hver av punktene. Husk:

```python
#plotter datasettet som kryss
plt.plot(tid,strekning,"xr")
parameter=np.polyfit(tid,strekning,4)
polynom=np.poly1d(parameter)

tidsverdier=np.linspace(0,46,200)
plt.plot(tidsverdier,polynom(tidsverdier))

plt.show()
print("polynomet som er tilpasset er P(X)=")
print(polynom)
```

## Momentan vekstfart
Så langt har vi regnet gjennomsnittsfart mellom to hekker, når vi nå har en funksjon som modell kan vi bruke den til å finne den momentane vekstfarten og derfor en modell på den momentane løpsfarten til Karstein. 

Når vi la h være veldig liten kan vi finne en tilnæringsverdi for den deriverte til en funksjon som:

$$f'(x)\approx \frac{f(x+h)-f(x)}{h}$$


<div class="alert alert-block alert-info">
<b>Utfordring! </b>Definer en funksjon som deriverer numerisk (bruker utrykket over med en liten h). Tegn grafen til den deriverte av polynomet du fant i sted for hele gyldighetsområdet av modellen. Du har nå laget en modell av farten til Karstein Warholm i gulløpet! </div>


```python
def Derivert(x,h=0.001):
    return (polynom(x+h)-polynom(x))/h

fart_modell=Derivert(tidsverdier)
plt.plot(tidsverdier,fart_modell)
plt.xlabel('tid/$s$')
plt.ylabel('fart/$m/s$')
plt.title('Mommentanfart - den deriverte')
plt.show()
```

```python
plt.plot(tid_midt,fart,label='Snittsfart mellom hekkene')
plt.plot(tidsverdier,fart_modell,label='Momentanfart vha. den deriverte' )
plt.xlabel('tid/$s$')
plt.ylabel('fart/$m/s$')
plt.title('Fart - 400m hekk') 
plt.legend()
plt.show() 
```
