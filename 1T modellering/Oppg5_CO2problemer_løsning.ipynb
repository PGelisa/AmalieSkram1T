## Del A: $CO_2$-utslipp i Norge. Totalutslipp mellom 1060 og 2020

I første del av oppgaven skal vi se på hvordan $CO_2$-utslippene i Norge har endret seg de siste 60 årene.

#### 1. Gitt de historiske dataene dere har fått for norske $CO_2$-utslipp; hvilken matematisk modell passer best til å uttrykke Norges $CO_2$-utslipp de siste 60 år?

For å kunne svare på dette spørsmålet må vi se på ulike typer regresjon av dataene, og se hvilken modell som passer best. Dette kan gjøres ved hjelp av Geogebra eller Python.

Importerer aktuelle biblioteker

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

Leser så inn dataene

```python
df=pd.read_csv('CO2utslippNorge_1960_2020.csv',sep=';')
```

```python
df
```

Sjekker hvordan dataene ser ut, og må rydde litt i verdiene - endre navn på en kolonne og slette en kolonne

```python
df.rename(columns = {'Norske samlete utslipp av CO2 fra olje, gass, kull, fakling, sementproduksjon':'Årstall'}, inplace = True)
```

```python
df
```

```python
del df['Unnamed: 2']
```

```python
df
```

```python
df.to_csv('norskeCO2utslipp.csv')
```

Nå ser vi at dataene har det vi trenger - årstall og totale $CO_2$-utslippene fra Norge i millioner tonn.

Definerer de to kolonnene som x- og y-data

```python
x_data=df['Årstall']
y_data=df['MtCO2']
```

Vi begynner med å plotte dataene, for å se hvilke funksjoner som kan være aktuelle

```python
plt.plot(x_data,y_data,'x')
```

Vi kan prøve med ulike funksjoner

```python
L1 = np.polyfit(x_data, y_data,1)
L2 = np.polyfit(x_data, y_data,2)
L3 = np.polyfit(x_data, y_data,3)
L4 = np.polyfit(x_data, y_data,4)
```

```python
linear = np.poly1d(L1)
andregrad = np.poly1d(L2)
tredjegrad = np.poly1d(L3)
fjerdegrad = np.poly1d(L4)
print(linear)
print(andregrad)
print(tredjegrad)
print(fjerdegrad)
```

```python
x_verdier = np.linspace(1972, 2022, 10000)
plt.plot(x_data,y_data,'x')
plt.plot(x_verdier, linear(x_verdier),label='lineær')
plt.plot(x_verdier, andregrad(x_verdier),label='andregrad')
plt.plot(x_verdier, tredjegrad(x_verdier),label='tredjegrad')
plt.plot(x_verdier, fjerdegrad(x_verdier),label='fjerdegrad')
plt.title('Årlige $CO_2$-utslipp i Norge')
plt.xlabel('Årstall')
plt.ylabel('MtCO2')
plt.grid()
plt.legend()
```

Hvis vi ser på de ulike kurvene, så kan de se ut til at fjernegradsfunksjonen passer best for det som har skjedd de siste 50 årene.

### 2. Hvis dere skal ta hensyn til mulige fremtidsscenarier; hvilken matematisk modell vil dere si passer best for å beskrive Norges $CO_2$-utslipp de siste 60 år OG som en fremtidsmodell (prognose)? 

```python
x_verdier = np.linspace(1960, 2030, 10000)
plt.plot(x_data,y_data,'x')
plt.plot(x_verdier, linear(x_verdier),label='lineær')
plt.plot(x_verdier, andregrad(x_verdier),label='andregrad')
plt.plot(x_verdier, tredjegrad(x_verdier),label='tredjegrad')
plt.plot(x_verdier, fjerdegrad(x_verdier),label='fjerdegrad')
plt.title('Årlige $CO_2$-utslipp i Norge')
plt.xlabel('Årstall')
plt.ylabel('MtCO2')
plt.grid()
plt.legend()
```

Hvis vi begynner med å se på hvordan modellene ser ut frem til 2030, året som er beskrevet i Paris-avtalen, så vet vi at Norge skal redusere utslippene med 55% innen 2030, som betyr at fjerdegradsfunksjonen kan fungere hvis vi klarer å redusere utslippene.

Men la oss se på hvordan modellen går frem til 2060:

```python
x_verdier = np.linspace(1960, 2060, 10000)
plt.plot(x_data,y_data,'x')
plt.plot(x_verdier, linear(x_verdier),label='lineær')
plt.plot(x_verdier, andregrad(x_verdier),label='andregrad')
plt.plot(x_verdier, tredjegrad(x_verdier),label='tredjegrad')
plt.plot(x_verdier, fjerdegrad(x_verdier),label='fjerdegrad')
plt.title('Årlige $CO_2$-utslipp i Norge')
plt.xlabel('Årstall')
plt.ylabel('MtCO2')
plt.grid()
plt.legend()
```

Vi ser at modellen vil fungere ganske dårlig når vi går lengre frem i tid.

## Del B: Atmosfærisk $CO_2$ 
Starter med å lese inn dataene

```python
xx_data = df2['Årstall']
yy_data = df2['Endringsrate (ppm/år)']
plt.plot(xx_data,yy_data,'x')
```

```python
deriv1 = np.poly1d(np.polyfit(xx_data,yy_data,1))
```

```python
deriv2 = np.poly1d(np.polyfit(xx_data,yy_data,2))
deriv3 = np.poly1d(np.polyfit(xx_data,yy_data,6))
```

```python
xx_verdier = np.linspace(1960,2030,10000)
plt.plot(xx_data,yy_data,'x')

plt.plot(xx_verdier,deriv1(xx_verdier),label='linje')
plt.plot(xx_verdier,deriv2(xx_verdier),label='parabel')
plt.plot(xx_verdier,deriv3(xx_verdier),label='n-te-grad')

plt.title('Årlig endring i globale $CO_2$-utslipp (ppm/år)')
plt.xlabel('Årstall')
plt.ylabel('ppm/år')
plt.grid()
plt.legend()
```
