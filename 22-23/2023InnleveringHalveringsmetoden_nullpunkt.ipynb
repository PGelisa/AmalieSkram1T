# Innlevering Halveringsmetoden - numerisk lsning av likninger
Hvordan kan vi finne nullpunkter dersom vi ikke kan lse en likning ved hjelp av det vi har lrt?

Tenk deg at vi tar utgangspunkt i et intervall $[a,b]$, se figuren under, som er slik at funksjonsverdien har forskjellig fortegn i endepunktene til intervallet (p ulik side av x-aksen). Da vet du at nullpunktet til funksjonen m vre i dette intervallet. Midtpunktet,$m_{1}$, beregnes og vi lager et nytt intervall ved  flytte enten a til $m_{1}$ eller b til $m_{1}$ slik at det nye intervallet $ [a_{2},b_{2}]$ ogs inneholder nullpunktet. Vi har n ftt et smalere intervall. Vi halverer intervallet igjen og ser p halvdelen som har forskjellig fortegn p funksjonsverdiene i endepunktet til det nye intervallet $ [a_{3},b_{3}]$. Dette fortsetter vi med til endepunktene er s nr hverandre som vi nsker. Endepunktene er da en tilnrming til nullpunktet.

Denne metoden finner bare et nullpunkt. Vi m derfor dele opp i to ulike intervall og kjre metoden to ganger om vi skal finne to nullpunkt for samme funksjon.

Vi m ogs "gjette" riktig intervall som nullpunktet befinner seg innenfor 







## 4.5 Eksempel: finne nullpunkt til en liner funksjon.

```python
# Vi skal lse likningen 3x-4=0. 
# Vi legger merke til 
# f(-2)=3*(-2)-4=-10
# f(2)=3*2-4=2
# Dermed har vi at f(-2) og f(2) ulikt fortegn for denne funksjonen
# og m passere null mellom disse
# Velger nedre grenser ut fra dette.


def f(x):
    return 3*x - 4 

nedregrense = -2
vregrense = 2
presisjon = 0.001

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
    if f(a) * f(m) < 0:
        b = m
    elif f(b) * f(m) < 0:  
        a = m
    elif f(m) == 0:
        break
    else:
        feil = True
        break
    m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.3f}.")
```

### Eksempel 4.5.5 Finne nullpunkt til sammensatte funksjoner.
Vi ser p likningen: $$2^x - 3x^3 = 0 , x\in[0,2]$$

Kan vi lse denne likningen?
Svaret er egentlig at vi ikke kan lse denne likningen ved hjelp av det vi har lrt i 1T uten numeriske lsninger. Det er derimot mulig  bruke halveringsmetoden.

```python
def f(x):
    return 2**x - 3*x**3  

nedregrense = 0
vregrense = 2
presisjon = 1e-4 # dette betyr det samme som 1*10^(-4)

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
    if f(a) * f(m) < 0:
        b = m
    elif f(b) * f(m) < 0:
        a = m
    elif f(m) == 0:
        break
    else:
        feil = True
        break
    m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.4f}.")
```

## 4.5 Eksempel 2: Finne flere nullpunkter ved halveringsmetoden
Vi m gjenta halveringsmetoden flere ganger med forskjellige intervaller dersom vi skal finne flere nullpunkt. Vi kan da "gjette" startverdi og sluttverdi til vi finner nullpunktene. Ofte kan det vre vanskelig  avgjre hvor mange nullpunkt en funksjon vil ha. Ved  tegne grafen til funksjonen vil vi kunne se hva som er fornuftige innverdier i halveringsmetoden. Kjr programmet under!

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1,1,1001) #Her velger vi x-verdier fra -1 til 1, jevnt forledt p 100 punkter.

def f(x): #definerer funksjonen f
    return( 2**x - 3*x**3 - 1)
plt.plot(x,f(x),label=r'$2^x - 3\cdot x^3 - 1$') #tegner grafen, markrer vises p hvert punkt (kan oftest droppes).
plt.axhline(0, color='green')
plt.grid(color='r', linestyle='-', linewidth=0.5)
plt.xlabel('x') #aksetittel x-akse
plt.ylabel('y') #aksetittel y-akse
plt.legend() #navn p funksjon
```

Vi ser av plottet at funksjonen har tre nullpunkt. Vi velger start -og sluttverdier utfra grafen.

Vi ser for eksempel at det frste nullpunktet ligger mellom $-0.50$ og $-0.25$, og velger dette som frste vre og nedre grense.

```python
def f(x):
    return( 2**x - 3*x**3 - 1) 

nedregrense = -0.5
vregrense = -0.25
presisjon = 1e-4 # dette betyr det samme som 1*10^(-4)

a = nedregrense
b = vregrense
m = (a + b)/2
feil = False

while b - a > presisjon:
    if f(a) * f(m) < 0:
        b = m
    elif f(b) * f(m) < 0:
        a = m
    elif f(m) == 0:
        break
    else:
        feil = True
        break
    m = (a + b)/2

if feil:
    print(f"f({a}) og f({b}) er p samme side av x-aksen.")
    print(f"Jeg finner ingen lsning.")
else:
    print(f"Funksjonen har et nullpunkt i x = {m:.4f}.")
```
