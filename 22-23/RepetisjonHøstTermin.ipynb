B
Lager tabell mens vi ser etter mønster

| Figurnummer | Verdi | Mønster |
| ----------- | ----- | ------- |
| 1 | 1 | |
| 2 | 3 | $F_1$ + 2 |
| 3 | 6 | $F_2$ + 3 |
| 4 | 10 | $F_3$ + 4 |
| n |  | $F_{n-1}$ + n |

| Figurnummer | Verdi | Formel |
| ----------- | ----- | ------- |
| 1 | 1 | $\dfrac{1\cdot 2}{2}$ |
| 2 | 3 | $\dfrac{2\cdot 3}{2}$ |
| 3 | 6 | $\dfrac{3\cdot 4}{2}$ |
| 4 | 10 | $\dfrac{4\cdot 5}{2}$ |
| n | | $\dfrac{n \cdot (n+1)}{2}$ |

```python
n = 10 # n er maks figurnummer
figurtall = 0
i = 0 # brukes til å telle opp alle figurtallene
for i in range(n):
    i = i + 1
    figurtall = figurtall + i
print(f"Figurnummer {i} er lik {figurtall}")
```

```python
i = 0 # starter med at vi ikke har noen figurer
figurtall = 0
sum = 0 # summen av alle prikkene vi har brukt
while sum <= 15000:
    i = i + 1 # figurnummeret øker med 1 hver gang 
    figurtall = figurtall + i # figurtallet øker med figurnummeret hver gang
    sum = sum + figurtall # summen av prikker som er brukt øker med figurtallet
print(f"Etter figurtall nummer {i} har vi brukt {sum} prikker")
print(f"Dette betyr at vi kan legge {i-1} figurer når vi har 15000 prikker")
```
