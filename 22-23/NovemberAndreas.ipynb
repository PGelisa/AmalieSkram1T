```python
import numpy as np
import matplotlib.pyplot as plt
 
def f(x):
    return 10000*x**(-1)
 
 
# x-verdiene
xs = np.linspace(1, 10, 1000)
# y-verdier
ys = f(xs)
 
# Tegne grafen
plt.plot(xs,ys, label="f(x)")
# Koordinatsystem
plt.grid()
# Forklaring x og y
plt.xlabel("Antall personer")
plt.ylabel("Kroner per person")
# Tittel
plt.title("Kostnad per person for å låne en hytte")
# Innholdsfortegnelse
plt.legend()
 
# Y-verdier til x=2 og x=5
K2 = f(2)
K5 = f(5)
 
 
print(f"Hvis 2 personer er med på turen koster det {K2} kroner per.")
print(f"Hvis 5 personer er med på turen koster det {K5} kroner per.")
 
 
# Tegnet inn i koordinatsystemet
plt.plot(2, K2, 'o')
plt.plot(5, K5, 'o')
```
