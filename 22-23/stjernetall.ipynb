Vi skal regne ut hvor mange stjernetall vi kan legge i flgende oppgave


Vi kan finne formelen for stjernetall ved regresjon i geogebra, eller vi kan se etter mnsteret i figurene:



Hvis vi velger  se etter mnsteret, s m vi kunne formelen for trekanttall: $T_i = \frac{i(i+1)}{2}$.

Figuren viser at stjernetallene kommer fra (en i midten) + seks trekanttall gitt ved (n-1) i heksagonene i midten + seks trekanttall gitt ved (n-1) p taggene.

Da fr vi: $S_n = 1 + 6\cdot \frac{(n-1)n}{2} + 6\cdot \frac{(n-1)n}{2} = 1 + 6\cdot (n-1)n = 1 + 6n^2 - 6n$

Formelen for stjernetall nummer n er gitt ved 
$$S_n = 6n^2 - 6n + 1$$

Vi skal se hvor mange stjerner vi kan legge med 1 million kuler eller prikker

```python
n = 0 # dette er telletallet, figurnummeret
stjerne = 0 # dette er antall kuler i stjernen
total = 0 # dette er totalsummen av kuler som er brukt

while total <= 1e6:
    n = n+1
    stjerne = 6*n**2 - 6*n + 1
    total = total + stjerne
    
print(f"Du kan legge {n-1} stjerner om du fr utdelt en million kuler.")
print(f"Da har du {1e6 - (total - stjerne)} kuler igjen som du ikke fr brukt.")
```
