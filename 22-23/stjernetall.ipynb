## Mønsteroppgave
Vi skal regne ut hvor mange stjernetall vi kan legge i følgende oppgave


Vi ser at formelen for stjernetall nummer n er gitt ved 
$$S_n = 6n^2 - 6n + 1$$

Vi skal se hvor mange stjerner vi kan legge med 1 million kuler eller prikker

```python
n = 0 # dette er telletallet, figurnummeret
stjerne = 0 # dette er antall kuler i stjernen
total = 0 # dette er totalsummen av kuler som er brukt

while total <= 1e6:
    n = n+1
    stjerne = 6*n**2 - 6*n + 1
    total = total + stjerne
    
print(f"Du kan legge {n-1} stjerner om du får utdelt en million kuler.")
print(f"Da har du {1e6 - (total - stjerne)} kuler igjen som du ikke får brukt.")
```
