## Notater fra undervisning til kapittel 4 funksjoner
I denne notatblokken finnes notatene fra ulike eksempler i undervisningen

Vi definerer funksjoner i Python med kommandoen `def`.

Under definerer vi eksempelvis en funksjon for mønsteret trekanttall

```python
def trekanttall(n): # inputverdien til funksjonen er n - her er det figurtallnummeret
    return n*(n+1)/2
```

Når vi har definert en funksjon, så kan vi bruke den senere

```python
trekanttall(10)
```

Vi kan definere funksjoner på samme måten:

```python
def f(x):
    return x**2 + x + 1/x
```

```python
f(-1)
```

### Vi kan bruke Python til å lage funksjonsplot
Da trenger vi et par biblioteker

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x

x_verdier = np.linspace(-5,5,1001) # her lager vi en array med x-verdier
# det blir som definisjonsmengden, og antall x-verdier som vi bruker i funksjonen
y_verdier = f(x_verdier) # regner ut tilhørende y-verdier

plt.plot(x_verdier,y_verdier)
plt.show()
```

Vi kan pynte på plottene med rutenett, aksekors mm

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2 + 2*x)/(x**2 + 1)

x_verdier = np.linspace(-5,5,1001)
y_verdier = f(x_verdier)

plt.xlabel("x")                 # aksetittel langs x-aksen
plt.ylabel("f(x)")              # aksetittel langs y-aksen
plt.axhline(y=0, color="black") # legger til en x-akse
plt.axvline(x=0, color="black") # legger til en y-akse
plt.grid()                      # legger til et rutenett

plt.plot(x_verdier,y_verdier,"g")
plt.show()
```

### Løser oppgave 4.89 i læreboken

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x

x_verdier = np.linspace(-5,5,100)
y_verdier = f(x_verdier)

plt.xlabel("x")                 # aksetittel langs x-aksen
plt.ylabel("f(x)")              # aksetittel langs y-aksen
plt.axhline(y=0, color="black") # legger til en x-akse
plt.axvline(x=0, color="black") # legger til en y-akse
plt.grid()                      # legger til et rutenett

plt.plot(x_verdier,y_verdier)
plt.show()

print(f"f(-5)={f(-5)}, f(-1)={f(-1)}, f(3)={f(3)}, f(5)={f(5)}")
print(f"Største y-verdi er {max(y_verdier)}")
print(f"Minste y-verdi er {min(y_verdier)}")
```

Vi ser at bunnpunktet er ved $x=-1$, og vi har et lokalt toppunkt når $x=5$.
Dette gir oss følgende verdimengde:
$$V_f = [-1, 35]$$

## Vi kan lage et fast program for å regne ut stigningstall til en linje gjennom to punkt

```python
def stigning(x1,x2,f): # vi må ha to x-verdier og en funksjon
    y1=f(x1)
    y2=f(x2)
    a = (y2-y1)/(x2-x1)
    return a
```

```python
a=stigning(0,4,f)
print(a)
```
