### Utforsk


```python
import numpy as np
import matplotlib.pyplot as plt

def W(x):
    return (-2.34*x**3 + 50*x**2 + 129*x + 19.7)

x_verdier = np.linspace(0,15,1001)
y_verdier = W(x_verdier)

plt.title("Antall artikler på Wikipedia")
plt.xlabel("x")                 # aksetittel langs x-aksen
plt.ylabel("W(x)")              # aksetittel langs y-aksen
plt.axhline(y=0, color="black") # legger til en x-akse
plt.axvline(x=0, color="black") # legger til en y-akse
plt.grid()                      # legger til et rutenett

plt.plot(x_verdier,y_verdier,"g")
plt.show()
```
