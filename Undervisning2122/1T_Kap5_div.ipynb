## Likningssystemer og ulikheter



```python
# Programmet lser et linert likningssystem med to
# ukjente nr ae  bd ikke er 0.
a = float(input("Skriv inn koeffisient a: "))
b = float(input("Skriv inn koeffisient b: "))
c = float(input("Skriv inn tallet c: "))
d = float(input("Skriv inn koeffisient d: "))
e = float(input("Skriv inn koeffisient e: "))
f = float(input("Skriv inn tallet f: "))

# Sjekker frst om ae - bd = 0
if a*e==b*d: # hvis dette stemmer, s har linjene likt stigningstall og er parallelle og kanskje sammenfallende
    if d/a == f/c: # dette betyr at de har likt konstantledd i tillegg til likt stigningstall
        print('Likningssettet har uendelig mange lsninger (to sammenfallende linjer)')
    else: # her har de da likt stigningstall og ulikt konstantledd, parallelle og ikke sammenfallende
        print('Likningssettet ingen lsninger, to parallelle linjer')
else: #linjene er ikke parallelle, og dermed ett unikt skjringspunkt og en lsning p likningssettet
    x = (c*e-b*f) / (a*e-b*d)
    y = (a*f-c*d) / (a*e-b*d)
    print("Lsningen p likningssettet er:")
    print("x =", x, "og y =", y)
```
