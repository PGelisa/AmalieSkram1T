```python
import numpy as np
aarstall=np.array([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])
lonn=np.array([502,538,552,592,657,618,635,619,640,748,751,810])
topplektor=np.array([518,527,555,559,583,592,607,619,631,651,653,675])
kpi=np.array([92.1,93.3,93.9,95.9,97.9,100,103.6,105.5,108.4,110.8,112.2,119.3])
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
plt.title("Lønnsutvikling siden 2010",color="g")
ax2=ax.twinx()
ax.plot(aarstall,lonn,color='r',label="Min lønn")
ax.set_ylabel("Total Lønn (1000 kr)",color="r",fontsize=14)
ax.plot(aarstall,topplektor,color="y",label="Lektor I, max ans")
ax2.plot(aarstall,(lonn/kpi)/(lonn[0]/kpi[0]),color="b",label="Min relative lønn")
ax2.plot(aarstall,(topplektor/kpi)/(topplektor[0]/kpi[0]),color="orange",label="Lektor I, max ans, relativ")
ax2.set_ylabel("Lønn/KPI i forhold til 2010",color="b",fontsize=14)
fig.legend(bbox_to_anchor=(1.4,1.), loc="upper right",borderaxespad=1) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x

x_verdier = np.linspace(-5,5,1001)
y_verdier = f(x_verdier)

plt.xlabel("x")                 # aksetittel langs x-aksen
plt.ylabel("f(x)")              # aksetittel langs y-aksen
plt.axhline(y=0, color="black") # legger til en x-akse
plt.axvline(x=0, color="black") # legger til en y-akse
plt.grid()                      # legger til et rutenett

plt.plot(x_verdier,y_verdier)
plt.show()
print(f"Største y-verdi er {max(y_verdier)}")
print(f"Minste y-verdi er {min(y_verdier)}")
```

```python
import random
import numpy as np
def antallKort():
    teller = 0
    kort_funnet = []
    while len(kort_funnet)<52:
        nyttKort = random.randint(1,52)
        if nyttKort not in kort_funnet:
            kort_funnet.append(nyttKort)
        teller +=1
    return teller, kort_funnet

statistikk=[]
antall_forsok = 1000000
for i in range(antall_forsok):
    teller, kort_funnet = antallKort()
    statistikk.append(teller)

print(np.mean(statistikk))
```
