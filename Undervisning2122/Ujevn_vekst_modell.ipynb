####  Oppgave 2 Vi bruker delighetstester for å løse problemer med ujevn vekst


Anikken trenger ny gaming PC og ønsker å spare 10 000 kr. Hun får 300 kr hver uke, 
men annenhver uke er hun hos moren sin i helgen og får 200 kr ekstra. 

Hun skriver et lite program for å regne ut hvor lenge hun må spare for å komme i mål:

```python
saldo =1000
uke=0
spare_mål=10000

saldo_oversikt=[saldo]
uke_oversikt=[uke]

while saldo < spare_mål:
        
    #øker saldo slik det skjer hver uke
    saldo = saldo + 300 
       
    if uke%2==0: #sjekker om det er en uke hun er hos moren
        saldo = saldo + 200 
     
    uke = uke + 1
    
    # Legger denne ukens saldo til bakerst i listen 
    saldo_oversikt.append(saldo)
    uke_oversikt.append(uke)
    
print("Det tar ",uke,"uker å spare mer en",spare_mål,"kr. Da er saldoen",saldo,"kr")
```

##### Hvor lang tid tar det Anikken å spare til ny PC? 

Anikken plotter listen med saldooversikt for å sjekke at resultatet er fornuftig:

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(uke_oversikt,saldo_oversikt)
```

Anikken syns dette tok lang tid. Så hver 4 uke hjelper hun farmor med storhandel og husvask. For dette arbeide får hun 1000 kr. Dessverre må hun også kjøpe nytt ungdomskort, så hver 5 uke må hun betale 370 kr. Gjør nødvendige forandringer på programmet for å hjelpe henne å finne ut hvor lang tid det tar å spare nå:

```python
saldo =1000
uke=0
spare_mål=10000

saldo_oversikt=[saldo]
uke_oversikt=[uke]

while saldo < spare_mål:
        
    #øker saldo slik det skjer hver uke
    saldo = saldo + 300 
       
    if uke%2==0: #sjekker om det er en uke hun er hos moren
        saldo = saldo + 200 
    
    if uke%4==0:
        saldo = saldo + 1000
    
    if uke%5==0:
        saldo = saldo - 370
     
    uke = uke + 1
    
    # Legger denne ukens saldo til bakerst i listen 
    saldo_oversikt.append(saldo)
    uke_oversikt.append(uke)
    
print("Det tar ",uke,"uker å spare mer en",spare_mål,"kr. Da er saldoen",saldo,"kr")
```

```python
# Skriv nødvendig kode for å plotte utviklingen på kontoen til Anikken. Kan du skrive på navn på aksene?
plt.plot(uke_oversikt,saldo_oversikt)
plt.xlabel("Uker")
plt.ylabel("Saldo")
plt.title("Anikkens spareplan")
plt.grid()
plt.show()
```
